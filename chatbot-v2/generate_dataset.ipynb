```python
import json
```

```python
# Read all names
with open("./datasets/dataset_cornell_movie/movie_characters_metadata.txt", "r") as f:
    lines = f.readlines()

characters = {}
for line in lines:
    cols = line.split(" +++$+++ ")
    characters[cols[0]] = cols[1].title()
characters
```

```python
# Read all names
with open("./datasets/dataset_cornell_movie/movie_conversations.txt", "r") as f:
    conv_lines = f.readlines()

with open("./datasets/dataset_cornell_movie/movie_lines.txt", "r") as f:
    full_lines = f.readlines()

all_turns = {}

for line in full_lines:
    cols = line.split(" +++$+++ ")
    all_turns[cols[0]] = (cols[3].title(), cols[4].strip().replace("  ", " ").replace("  ", " ").replace("  ", " ").replace("  ", " "))

conversations = []

failed = 0
for line in conv_lines:
    try:
        cols = line.split(" +++$+++ ")
        char1, char2 = characters[cols[0]], characters[cols[1]]
        turns = [all_turns[t] for t in json.loads(cols[3].replace("'", '"'))]
        conversations.append(((char1, char2), turns))
    except:
        failed += 1
print(failed)
```

```python
with open("./datasets/cornell_parsed.json", "w") as f:
    json.dump(conversations, f)
```
